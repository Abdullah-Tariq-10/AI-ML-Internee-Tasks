```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="ticks")

print("--- 1. Loading and Inspecting Data ---")

iris_df = sns.load_dataset('iris')

print("\nDataset Shape:")
print(iris_df.shape)

print("\nColumn Names:")
print(iris_df.columns)

print("\nFirst 5 Rows (Head):")
print(iris_df.head())

```

```python
print("\n--- 2. Summary Statistics ---")

print("\nData Info:")
iris_df.info()

print("\nDescriptive Statistics:")
print(iris_df.describe())
```

```python
print("\n--- 3. Generating Visualizations ---")

print("Displaying Pairplot (Scatter Plots)...")
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Features by Species', y=1.02) 
plt.show()
```

```python
print("Displaying Histograms...")

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Histograms of Iris Feature Distributions', y=1.02)

sns.histplot(iris_df['sepal_length'], kde=True, ax=axes[0, 0]).set_title('Sepal Length')
sns.histplot(iris_df['sepal_width'], kde=True, ax=axes[0, 1]).set_title('Sepal Width')
sns.histplot(iris_df['petal_length'], kde=True, ax=axes[1, 0]).set_title('Petal Length')
sns.histplot(iris_df['petal_width'], kde=True, ax=axes[1, 1]).set_title('Petal Width')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```

```python
print("Displaying Box Plots...")

melted_df = pd.melt(iris_df, "species", var_name="feature", value_name="measurement")

plt.figure(figsize=(12, 8))
sns.boxplot(x="feature", y="measurement", hue="species", data=melted_df)
plt.title('Box Plots of Iris Features by Species')
plt.show()

```
